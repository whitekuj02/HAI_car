```python
import torch
import numpy as np

def cutout(image, size=50):
    """
    이미지의 정사각형 영역을 무작위로 잘라 검은색으로 만듭니다.
    Args:
        image: numpy array (H, W, C)
        size: 마스킹할 정사각형 한 변의 길이
    Returns:
        masked image
    """
    H, W, _ = image.shape
    x = np.random.randint(W)
    y = np.random.randint(H)

    x1 = np.clip(x - size // 2, 0, W)
    x2 = np.clip(x + size // 2, 0, W)
    y1 = np.clip(y - size // 2, 0, H)
    y2 = np.clip(y + size // 2, 0, H)

    image_copy = image.copy()
    image_copy[y1:y2, x1:x2, :] = 0  # ⚠️ HWC 순서
    return image_copy

def block_masking(image, num_blocks=20, block_size=20):
    """
    이미지 위에 랜덤한 위치에 작은 검정 네모 박스를 여러 개 덮음.
    Args:
        image: numpy array (H, W, C)
        num_blocks: 블록 개수
        block_size: 블록 한 변 크기
    Returns:
        masked image
    """
    image_copy = image.copy()
    H, W, _ = image.shape

    for _ in range(num_blocks):
        x = np.random.randint(0, W - block_size)
        y = np.random.randint(0, H - block_size)
        image_copy[y:y+block_size, x:x+block_size] = 0  # 검정 박스

    return image_copy

```

```python
image_path = "/home/aicontest/HAI_car/data/train/1시리즈_F20_2013_2015/1시리즈_F20_2013_2015_0000.jpg"

import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

train_transform = A.Compose([
    # ✅ 정보 손실을 줄이고, 세부 crop 보존
    A.RandomResizedCrop(
        size=(224, 224),
        scale=(0.5, 1.0),       # 50~100% 면적
        ratio=(0.75, 1.33),     # 더 다양한 aspect ratio
        p=1.0
    ),

    # ✅ 너무 센 augment 제거하고, 디테일 보존 위주로 선택
    A.SomeOf([
        A.HorizontalFlip(p=1.0),  # 대칭만 유지
        A.RandomBrightnessContrast(p=1.0),  # 전체 명도 대비 조정
        A.HueSaturationValue(p=1.0),  # 색조 변경 (너무 세지 않게)
        A.CLAHE(p=1.0),  # 국소 대비 강화 → fine-grained에 도움
    ], n=2, replace=False, p=0.8),  # 2개만 선택

    A.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225])
])

image = train_transform(image=image)['image']

# 시각화
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("transform")
plt.axis('off')

cutout_image = cutout(image)

plt.subplot(1, 3, 2)
plt.imshow(cutout_image)
plt.title("CutOut")
plt.axis('off')

salt_and_pepper_image = block_masking(image)

plt.subplot(1, 3, 3)
plt.imshow(salt_and_pepper_image)
plt.title("Salt & Pepper")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
test_transform = A.RandomResizedCrop(
    size=(224, 224),
    scale=(0.5, 1.0),       # 50~100% 면적
    ratio=(0.75, 1.33),     # 더 다양한 aspect ratio
    p=1.0
)

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = test_transform(image=image)['image']
plt.figure(figsize=(15, 5))
plt.subplot(1, 1, 1)
plt.imshow(image)
plt.title("transform")
plt.axis('off')
```
